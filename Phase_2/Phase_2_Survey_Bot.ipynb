##### STT code

```python
import tempfile
import torch
import os
import numpy as np
from scipy.io.wavfile import write
import logging
import speech_recognition as sr
from pydub import AudioSegment
from transformers import AutoModelForSeq2SeqLM, BitsAndBytesConfig, AutoTokenizer
from IndicTransToolkit import IndicProcessor
import threading
import sys
import time

# Configure logging
logging.basicConfig(level=logging.INFO)

# Load IndicTrans model and tokenizer
MODEL_NAME = "ai4bharat/indictrans2-indic-en-1B"
config = BitsAndBytesConfig(load_in_8bit=True)
indic_en_tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
indic_en_model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME, trust_remote_code=True)
ip = IndicProcessor(inference=True)

# Global variables to store output
stored_kannada_text = None
stored_english_translation = None

# ✅ Convert audio to WAV format using pydub
def convert_to_wav(audio_file):
    try:
        audio = AudioSegment.from_file(audio_file)
        temp_wav_file = tempfile.NamedTemporaryFile(delete=False, suffix=".wav")
        audio.export(temp_wav_file.name, format="wav")
        return temp_wav_file.name
    except Exception as e:
        logging.error(f"Error converting file to WAV: {e}")
        return None

# ✅ Function to transcribe Kannada audio
def transcribe_audio(audio_file):
    try:
        logging.info(f"Converting audio file: {audio_file}")
        converted_file = convert_to_wav(audio_file)
        if not converted_file:
            return "Error converting audio file to WAV"

        recognizer = sr.Recognizer()
        with sr.AudioFile(converted_file) as source:
            audio = recognizer.record(source)

        logging.info("Running SpeechRecognition for transcription")
        transcription = recognizer.recognize_google(audio, language="kn-IN")
        logging.info(f"Transcription result: {transcription}")

        os.remove(converted_file)  # Remove temporary file after processing
        return transcription.strip()
    except sr.UnknownValueError:
        logging.error("SpeechRecognition could not understand the audio")
        return "Could not understand the audio"
    except sr.RequestError as e:
        logging.error(f"SpeechRecognition service error: {e}")
        return f"SpeechRecognition error: {e}"
    except Exception as e:
        logging.error(f"Error during transcription: {e}")
        return f"Error: {e}"

# ✅ Function to translate Kannada to English using IndicTrans
def translate_sentence(sentence):
    try:
        src_lang, tgt_lang = "kan_Knda", "eng_Latn"
        batch = [sentence]

        # Preprocess input
        batch = ip.preprocess_batch(batch, src_lang=src_lang, tgt_lang=tgt_lang)

        # Tokenize input
        inputs = indic_en_tokenizer(
            batch,
            truncation=True,
            padding="longest",
            return_tensors="pt",
            return_attention_mask=True,
        )

        # Generate translation
        with torch.no_grad():
            generated_tokens = indic_en_model.generate(
                **inputs,
                use_cache=True,
                min_length=0,
                max_length=256,
                num_beams=5,
                num_return_sequences=1,
            )

        # Decode output
        with indic_en_tokenizer.as_target_tokenizer():
            generated_tokens = indic_en_tokenizer.batch_decode(
                generated_tokens.detach().cpu().tolist(),
                skip_special_tokens=True,
                clean_up_tokenization_spaces=True,
            )

        # Post-process result
        translations = ip.postprocess_batch(generated_tokens, lang=tgt_lang)
        return translations[0] if translations else "Translation failed."
    except Exception as e:
        logging.error(f"Error during translation: {e}")
        return f"Error: {e}"

def transcribe_audio_from_mic(language='kn-IN'):
    recognizer = sr.Recognizer()
    recognizer.dynamic_energy_threshold = True
    recognizer.pause_threshold = 1.0
    
    stop_event = threading.Event()

    with sr.Microphone() as source:
        print("Recording... Speak now!")
        recognizer.adjust_for_ambient_noise(source)

        audio_data = []
        start_time = time.time()  # Start time tracking

        def stop_recording():
            input("Press Enter to stop recording...\n")
            stop_event.set()

        stop_thread = threading.Thread(target=stop_recording)
        stop_thread.start()

        try:
            while not stop_event.is_set():
                audio = recognizer.listen(source)
                audio_data.append(audio)

            end_time = time.time()  # End time tracking
            duration = round(end_time - start_time, 2)  # Calculate duration in seconds

            print(f"Recording duration: {duration} seconds")

            if not audio_data:
                print("No audio captured.")
                return None, duration

            # Combine the audio chunks
            combined_audio = sr.AudioData(
                b''.join([a.get_raw_data() for a in audio_data]),
                source.SAMPLE_RATE,
                source.SAMPLE_WIDTH
            )

            # Attempt transcription
            try:
                text = recognizer.recognize_google(combined_audio, language=language)
                print("Transcription:", text)
                return text, duration
            except sr.UnknownValueError:
                print("Speech Recognition could not understand the audio.")
                return None, 0
            except sr.RequestError as e:
                print(f"Could not request results: {e}")
                return None, 0
        except Exception as e:
            print(f"Error: {e}")
            return None, 0
        finally:
            stop_event.set()
            stop_thread.join()

# ✅ Function to process input
def process_input(input_type, audio_file=None, text_input=None):
    if input_type == "1" and audio_file:
        kannada_text = transcribe_audio(audio_file)
    elif input_type == "2":
        transcription, duration = transcribe_audio_from_mic(language='kn-IN')

        if transcription:
            kannada_text = transcription
        else:
            return "Could not transcribe live audio", "No translation"
        
    elif input_type == "3" and text_input:
        kannada_text = text_input
    else:
        return "Invalid input", "Invalid input"
    
    # Ensure that kannada_text is a string (not a tuple)
    if isinstance(kannada_text, tuple):
        kannada_text = kannada_text[0]

    english_translation = translate_sentence(kannada_text)
    return kannada_text, english_translation

# ✅ Save output to file
def save_output_to_file(kannada_text, english_translation, file_path="output.txt"):
    try:
        with open(file_path, "w", encoding="utf-8") as f:
            f.write(f"Kannada Text:\n{kannada_text}\n\n")
            f.write(f"English Translation:\n{english_translation}\n")
        logging.info(f"Output saved to {file_path}")
    except Exception as e:
        logging.error(f"Error saving output to file: {e}")

# ✅ Terminal-based input handling
def main():
    global stored_kannada_text, stored_english_translation
    
    print("\nSelect input type:")
    print("1. Audio File")
    print("2. Live Audio")
    print("3. Text Input")
    choice = input("Enter your choice (1/2/3): ").strip()

    if choice == "1":
        file_path = input("Enter path to the audio file: ").strip()
        if not os.path.exists(file_path):
            logging.error(f"Invalid file path: {file_path}")
            print("File not found. Please check the path and try again.")
            return
        kannada_text, english_translation = process_input("1", audio_file=file_path)

    elif choice == "2":
        logging.info("Starting live audio capture...")
        kannada_text, english_translation = process_input("2")

    elif choice == "3":
        text_input = input("Enter Kannada text: ").strip()
        kannada_text, english_translation = process_input("3", text_input=text_input)

    else:
        print("Invalid choice.")
        return

    # Store results globally for later use
    stored_kannada_text = kannada_text
    stored_english_translation = english_translation

    # print("\n--- Results ---")
    # print(f"Kannada Text: {kannada_text}")
    # print(f"English Translation: {english_translation}")

    # Save output to file
    save_output_to_file(kannada_text, english_translation)

if __name__ == "__main__":
    main()

    # Display stored results for debugging or later use
    print("\n--- Stored Results ---")
    print(f"Stored Kannada Text: {stored_kannada_text}")
    print(f"Stored English Translation: {stored_english_translation}")
```

```python
import os
import json
import re
from datetime import datetime
from datetime import time
import pandas as pd
import spacy
from transformers import pipeline

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Function to extract the most relevant keyword (preferably a verb)
def extract_main_keyword(question):
    doc = nlp(question.lower())
    
    # Extract verbs (action words) first, filtering out stopwords and punctuation
    verbs = [token.text for token in doc if token.pos_ == "VERB" and not token.is_stop and not token.is_punct]
    
    # If no verbs found, fall back to nouns (identity-related words), filtering out stopwords and punctuation
    if verbs:
        return verbs[0]  # Return the first verb found
    else:
        nouns = [token.text for token in doc if token.pos_ == "NOUN" and not token.is_stop and not token.is_punct]
        return nouns[0] if nouns else None  # Return first noun if exists, else None

# Define file paths
INPUT_FILE_PATH = os.path.expanduser("/Users/admin/ai-questionnaire-project/phase_2_qns_demo.xlsx")
#OUTPUT_FILE_PATH = os.path.expanduser("/Users/admin/ai-questionnaire-project/completed_qns.xlsx")

VOCAB_FILE = "Final_vocab_dict.json"

def load_or_create_vocab():
    default_vocab = {
        "Options_yes_no": {
            "valid_options": ["Yes", "No"],
            "Yes": ["yes", "i do", "i drink", "i have","a week"],
            "No": ["no", "i don't", "don't", "i do not","do not", "never"]
        },
        "Options_frequency": {
            "valid_options": ["Daily", "Weekly", "Monthly", "Never"],
            "Daily": ["daily", "a day", "every day"],
            "Weekly": ["weekly", "a week", "every week"],
            "Monthly": ["monthly", "a month", "once a month"],
            "Never": ["never", "not at all"]
        },
        "Options_gender": {
            "valid_options": ["Male", "Female", "Others"],
            "Male": ["male", "man", "boy", "gentleman", "he", "his", "him"],
            "Female": ["female", "woman", "girl", "lady", "her", "she", "hers"],
            "Others": ["other", "others", "they", "them", "non-binary"]
        },
        "Options_marital_status": {
            "valid_options": ["Single", "Married", "Divorced", "Widowed"],
            "Single": ["single", "unmarried", "not married", "bachelor", "single status"],
            "Married": ["married", "wife", "husband", "spouse", "partner", "in a relationship","engaged","getting married"],
            "Divorced": ["divorced", "separated", "no longer married"],
            "Widowed": ["widowed", "lost spouse", "spouse deceased", "widower", "widow"],
        },
        "Options_living_arrangement": {
            "valid_options": ["Alone", "With family", "Assisted living"],
            "Alone": ["alone", "by myself", "live alone", "independent", "on my own"],
            "With family": ["with family","family", "with parents", "with spouse", "with children", "with relatives"],
            "Assisted living": ["assisted living", "care home", "nursing home", "old age home", "retirement facility"]
        },
        "Options_addiction_frequency": {
            "valid_options": ["Daily", "2-3times a week", ">4 times a month", "Occasionally"],
            "Daily": ["daily", "every day", "once a day", "smoke daily", "drink daily", "chew daily"],
            "2-3times a week": ["2 times a week","3 times a week","2 days a week","3 days a week" ,"2 times","3 times", "few times a week","weekly", "couple times a week"],
            ">4 times a month": ["4 times a week","4 days a week","5 times a week","5 days a week","more than four times", "frequently in a month", "more than weekly", "more than four times a month"],
            "Occasionally": ["occasionally", "sometimes", "rarely", "once in a while"]
        },
        "Options_daily_fruit_servings": {
            "valid_options": ["0", "1", "2-3", "More Than 4"],
            "0": ["no", "i don't", "i do not", "never","0", "none", "no fruit"],
            "1": ["1", "single", "once", "just one", "only one"],
            "2-3": ["2", "3", "2 to 3", "two to three", "a couple", "few"],
            "More Than 4": ["more than 4", "5", "6", "many", "a lot", "more than 4", "four plus","4 plus"]
        },
        "Options_daily_water_intake": {
            "valid_options": ["< 1 L", "1-1.9 L", "2-2.9 L", "3-3.9 L", "More Than 4 L"],
            "< 1 L": ["less than 1", "under one", "half litre", "500 ml", "not even a litre", "small amount of water"],
            "1-1.9 L": ["1 litre", "1.5 litres", "between 1 and 2 litres", "1 to 2 litres"],
            "2-2.9 L": ["2 litres", "2.5 litres", "between 2 and 3 litres", "2 to 3 litres"],
            "3-3.9 L": ["3 litres", "three litres", "3.5 litres", "between 3 and 4 litres", "3 to 4 litres", "three to four litres"],
            "More Than 4 L": ["more than 4", "above 4", "5 litres", "drink a lot of water", "more than enough water", "four plus litres", "more than 4 litres"]
        },
         "Options_breakfast_days_per_week": {
            "valid_options": ["0", "1-2", "3-4", "More Than 5 days"],
            "0": ["never", "i don't", "not at all", "zero", "0", "skip every day"],
            "1-2": ["1", "2", "once or twice", "one or two", "rarely", "hardly", "occasionally", "few days"],
            "3-4": ["3", "4", "three to four", "moderately", "sometimes", "some days"],
            "More Than 5 days": ["5", "6", "7", "every day", "daily", "more than 5", "six or seven", "all week"]
        },
        "Options_monthly_oil_consumption": {
            "valid_options": ["< 250 ml", "250-500 ml", "500-1000 ml", "More Than 1 L"],
            "< 250 ml": ["less than 250", "under 250", "below 250", "very little", "< 250", "tiny amount","100","200","50","25"],
            "250-500 ml": ["250", "300","350", "400","450", "around 250", "around 500", "250 to 500", "between 250 and 500"],
            "500-1000 ml": ["500","550","600","650","700", "750", "800","850","900","950","1000", "up to 1 litre", "half to one liter", "500 to 1000", "between 500 and 1 litre"],
            "More Than 1 L": ["more than 1 litre", "over 1 litre", "above 1 litre", "1.5 litres", "2 litres", "a lot", "heavy consumption", "more than 1000 ml"]
        },
        "Options_pulses_vs_starch": {
            "valid_options": ["Almost Always", "Often", "Occasionally", "Rarely/Never"],
            "Almost Always": ["always", "every time", "consistently", "at least equal", "all the time", "without fail"],
            "Often": ["often", "most of the time", "usually", "regularly", "frequently"],
            "Occasionally": ["sometimes", "occasionally", "at times"],
            "Rarely/Never": ["rarely", "hardly", "never", "not usually", "seldom"]
        },
        "Options_sleep_time": {
            "valid_options": ["Before 7:00 PM", "By 8:30 PM", "By 10:00 PM", "After 10:00 PM"]
        },
        "Options_meal_count": {
            "valid_options": ["2 or less", "3", "4-5", "More Than 6"],
            "2 or less": ["1", "2", "just 1", "only 2", "a couple", "2 or less", "rarely eat", "very few"],
            "3": ["3", "exactly 3", "usually 3"],
            "4-5": ["4", "5", "4-5","4 or 5", "5 meals"],
            "More Than 6": ["6","more than 6", "above 6", "over 6", "7", "8", "9", "10", "a lot", "many meals"]
        },
        "num_mapping": {
            "one": "1", "two": "2", "three": "3", "four": "4", "five": "5",
            "six": "6", "seven": "7", "eight": "8", "nine": "9", "ten": "10",
            "once": "1", "twice": "2", "thrice": "3"
        }
    }

    if os.path.exists(VOCAB_FILE):
        with open(VOCAB_FILE, "r") as f:
            vocab = json.load(f)
    else:
        vocab = {}

    updated = False

    # Merge defaults into existing vocab
    for key, default_value in default_vocab.items():
        if key not in vocab:
            vocab[key] = default_value
            updated = True
        else:
            for sub_key, sub_val in default_value.items():
                if sub_key not in vocab[key]:
                    vocab[key][sub_key] = sub_val
                    updated = True

    if updated:
        with open(VOCAB_FILE, "w") as f:
            json.dump(vocab, f, indent=4)

    return vocab

vocab = load_or_create_vocab()

# Load Hugging Face QA model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# # Load questionnaire from Excel
def load_questionnaire(file_path):
    try:
        df = pd.read_excel(file_path)
        return df
    except Exception as e:
        print(f"Failed to load questionnaire. Error: {e}")
        return None


# Convert words to numbers using num_mapping
def convert_numbers(answer):
    return " ".join(vocab["num_mapping"].get(word, word) for word in answer.split())


def should_extract_answer(question, transcription):
    transcription = transcription.lower()
    question = question.lower()
    main_keyword = extract_main_keyword(question)

    keyword_map = {
        "gender": [
            "male", "man", "boy", "gentleman", "he", "his", "him",
            "female", "woman", "girl", "lady", "her", "she", "hers"
        ],
        "status": [
            "single", "unmarried", "not married", "bachelor",
            "married", "wife", "husband", "spouse", "partner", "engaged", "in a relationship",
            "divorced", "separated", "no longer married",
            "widowed", "widower", "widow", "lost spouse", "spouse deceased"
        ],
        "living": [
            "alone", "by myself", "live alone", "independent", "on my own",
            "with family", "family", "with parents", "with spouse", "with children", "with relatives",
            "assisted living", "care home", "nursing home", "old age home", "retirement facility"
        ],
        "drink": [
            "alcohol", "beer", "wine", "liquor", "whiskey",
            "tea", "coffee", "milk", "juice", "water"
        ],
        "eat": [
            "fruit", "fruits", "fresh fruit", "vegetable", "vegetables", "junk food", "snacks", "fish","breakfast","pulses","legumes","starchy foods","fats","last meal","last snack","tobacco"
        ],
        "consume": [
            "oil","meals plus snacks","meals"
        ],
        "chew": [
            "tobacco", "pan", "gutka", "betel", "betel nut"
        ]
    }

    if main_keyword in keyword_map:
        keywords = keyword_map[main_keyword]

        # For "drink", "eat", "chew" — check if the keyword is in both question and transcription
        if main_keyword in ["drink", "eat", "chew","consume"]:
            # Iterate through each keyword in the list
            for k in keywords:
                # Check if the current keyword is in both the question and transcription
                if k in question and k in transcription:
                    return True  # Return True as soon as a keyword matches both
            # If no keyword matches both, return False
            return False
    
        return any(k in transcription for k in keywords)

    # Fallback direct keyword presence
    return main_keyword and main_keyword in transcription

# # Extract answer using Hugging Face model
def extract_answer_with_model(question, transcription):
    try:
        transcription = transcription.lower()

        # Custom handling for fruit servings
        if "fruit" in question:
            # Add robust keyword checks for zero servings
            zero_keywords = ["don't eat fruit", "do not eat fruit", "no fruit", "never eat fruit", "zero fruit"]
            if any(phrase in transcription for phrase in zero_keywords):
                return "0"

        # Fallback to QA model
        result = qa_pipeline(question=question, context=transcription)
        return result['answer'].strip()

    except Exception as e:
        print(f"Model failed to extract answer: {e}")
        return None

# Time extraction function
def extract_time_from_text(text):
    # Check for words like "morning", "afternoon", "evening", etc., and map them to respective time ranges.
    time_of_day_mapping = {
        "morning": (6, 12),   # 6 AM to 12 PM
        "afternoon": (12, 18),  # 12 PM to 6 PM
        "evening": (18, 24),  # 6 PM to 12 AM
        "night": (0, 6)       # 12 AM to 6 AM
    }

    # Check if the transcription contains any of these words and set the appropriate time range
    for period, (start, end) in time_of_day_mapping.items():
        if period in text.lower():
            # Check if there's a number after the period (e.g., "evening 8")
            match = re.search(r'(\d{1,2})', text)
            if match:
                hour = int(match.group(1))
                if period == "evening":
                    # Evening + hour (like "evening 8") → 8 PM (20:00)
                    return datetime.strptime(f"{hour + 12}:00", "%H:%M").time()
                elif period == "morning":
                    # Morning + hour (like "morning 8") → 8 AM (08:00)
                    return datetime.strptime(f"{hour}:00", "%H:%M").time()
                elif period == "afternoon":
                    # Afternoon + hour (like "afternoon 8") → 8 PM (20:00)
                    return datetime.strptime(f"{hour + 12}:00", "%H:%M").time()
                elif period == "night":
                    # Night + hour (like "night 8") → 8 PM (20:00)
                    return datetime.strptime(f"{hour + 12}:00", "%H:%M").time()
            else:
                # If there's no hour mentioned, default to the start time of the period
                return datetime.strptime(f"{start}:00", "%H:%M").time()

    # Search for numeric time (with or without AM/PM)
    match = re.search(r'(\d{1,2})(?::\d{2})?\s*(am|pm)?', text.lower())
    
    # If no match is found, assume the time is 11 AM by default
    if not match:
        return datetime.strptime("11:00", "%H:%M").time()

    # Extract hour and AM/PM info
    hour = int(match.group(1))
    meridian = match.group(2)

    # Adjust hour based on AM/PM
    if meridian == "pm" and hour < 12:
        hour += 12
    elif meridian == "am" and hour == 12:
        hour = 0

    return datetime.strptime(f"{hour}:00", "%H:%M").time()

# Option classifier
def map_to_option(time_obj, key):
    if not time_obj:
        return None

    time_val = time_obj.hour + time_obj.minute / 60.0

    # Mapping based on cutoffs
    if time_val < 19:
        return "Before 7:00 PM"
    elif time_val <= 20.5:
        return "By 8:30 PM"
    elif time_val <= 22:
        return "By 10:00 PM"
    else:
        return "After 10:00 PM"


# Process the predicted answer
def process_answer(predicted_answer, options,vocab):
    
    if not predicted_answer:
        return "No Answer Found"
    
    answer = convert_numbers(predicted_answer.lower().strip())

    if options == "-":
        for key in ["No", "Yes"]:
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in vocab["Options_yes_no"][key]):
                return key
            
    if options == ["Yes","No"]:
        for key in ["No", "Yes"]:
            if key == "valid_options":
                continue  # Skip metadata            
            if any(ans in answer for ans in vocab["Options_yes_no"][key]):
                return key
        return "❗wrong answer"
    
    if options == ["Daily","Weekly","Monthly","Never"]:
        for key, values in vocab["Options_frequency"].items():
            if key == "valid_options":
                continue  # Skip metadata
            # print(answer,key,values)
            if any(ans in answer for ans in values):
                # print("key is",key)
                return key
        return "❗wrong answer"

    if options == ["Male","Female","Others"]:
        for key, values in vocab["Options_gender"].items():
            if key == "valid_options":
                continue  # Skip metadata
            # print(answer, key, values)  # Debugging statement to track what is being processed
            for ans in values:
                # print("key issss", key)
                if re.search(rf"\b{re.escape(ans)}\b", answer):
                    # print("key is", key)
                    return key
        return "❗wrong answer"

    if options == ["Single","Married","Divorced","Widowed"]:
        for key, values in vocab["Options_marital_status"].items():
            if key == "valid_options":
                continue  # Skip metadata
            # print(answer, key, values)  # Debugging statement to track what is being processed
            if any(ans in answer for ans in values):
                # print("Matched marital status:", key)
                return key
        return "❗wrong answer"
    
    if options == ["Alone", "With family", "Assisted living"]:
        for key, values in vocab["Options_living_arrangement"].items():
            if key == "valid_options":
                continue  # Skip metadata
            # print(answer, key, values)  # Debugging
            if any(ans in answer for ans in values):
                # print("Matched living arrangement:", key)
                return key
        return "❗wrong answer"

    if options == ["Daily", "2-3times a week", ">4 times a month", "Occasionally"]:
        for key, values in vocab["Options_addiction_frequency"].items():
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in values):
                # print("Matched addiction frequency:", key)
                return key
        return "❗wrong answer"
    
    if options == ["0", "1", "2-3", "More Than 4"]:
        for key, values in vocab["Options_daily_fruit_servings"].items():
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in values):
                # print("Matched addiction frequency:", key)
                return key
        return "❗wrong answer"
    
    if options == ["< 1 L", "1-1.9 L", "2-2.9 L", "3-3.9 L", "More Than 4 L"]:
        for key, values in vocab["Options_daily_water_intake"].items():
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in values):
                # print("Matched addiction frequency:", key)
                return key
        return "❗wrong answer"
    
    if options == ["0", "1-2", "3-4", "More Than 5 days"]:
        for key, values in vocab["Options_breakfast_days_per_week"].items():
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in values):
                # print("Matched addiction frequency:", key)
                return key
        return "❗wrong answer"
    
    if options == ["< 250 ml", "250-500 ml", "500-1000 ml", "More Than 1 L"]:
        for key, values in vocab["Options_monthly_oil_consumption"].items():
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in values):
                # print("Matched addiction frequency:", key)
                return key
        return "❗wrong answer"
    
    if options == ["Almost Always", "Often", "Occasionally", "Rarely/Never"]:
        for key, values in vocab["Options_pulses_vs_starch"].items():
            # print("check",key,values)
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in values):
                # print("Matched addiction frequency:", key)
                return key
        return "❗wrong answer"
    
    if options == ["Before 7:00 PM", "By 8:30 PM", "By 10:00 PM", "After 10:00 PM"]:
        time_obj = extract_time_from_text(predicted_answer)
        if time_obj:
            final_option = map_to_option(time_obj, options)
            # print("✅ Mapped Option:", final_option)
            return final_option
        return "❗wrong answer"  

    if options == ["2 or less", "3", "4-5", "More Than 6"]:
        for key, values in vocab["Options_meal_count"].items():
            # print("check",key,values)
            if key == "valid_options":
                continue  # Skip metadata
            if any(ans in answer for ans in values):
                # print("Matched addiction frequency:", key)
                return key
        return "❗wrong answer"

    return answer

def is_known_option_set(options, vocab):
    vocab_option_sets = [
        set(category["valid_options"])
        for key, category in vocab.items()
        if key.startswith("Options_")
    ]
    options_set = set(options)
    return any(options_set.issubset(valid_set) for valid_set in vocab_option_sets)

# Fill answers in the questionnaire
def fill_answers(df, transcription,vocab):
    for index, row in df.iterrows():
        qn_id=row['QN_ID']
        question = row['DESCRIPTION']
        question_type = row['Question_Type']
        options = row['Options']
        if options!='-':
            options=options.split(",")
        answer = "No Answer Found"

        if question_type == "Multiple Choice" and not is_known_option_set(options, vocab):
            df.at[index, 'Answer'] = "❗Engineer intervention required."
            print(options)
            continue # Skip further processing and go to the next question
        
        main_keyword = extract_main_keyword(question)
        print("qns_id--",qn_id,"main_keyword from qn--",main_keyword)
        
        if should_extract_answer(question, stored_english_translation):
            predicted_answer = extract_answer_with_model(question, stored_english_translation)
            print("qn--",question, ",answer--",predicted_answer, ",options--",options)
            processed_answer = process_answer(predicted_answer, options, vocab)
            answer = processed_answer

        df.at[index, 'Answer'] = answer
    
    return df


# Load questionnaire
df = load_questionnaire(INPUT_FILE_PATH)
if df is not None:
    df['Answer'] = "No Answer"
    updated_df = fill_answers(df, stored_english_translation,vocab)
    print("Transcription-",stored_english_translation)
    print("Final results-")
    df.fillna('', inplace=True)
    display(df)
```
